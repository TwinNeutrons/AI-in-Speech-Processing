```python
import librosa
import librosa.display
import IPython.display as ipd

y, sr = librosa.load("AI in Speech Processing.mp3")
librosa.display.waveshow(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
FFT_Signal = np.fft.fft(y)
plt.plot(FFT_Signal.real) #Plotting the amplitude of the signal
```

```python
IFFT_Signal = np.fft.ifft(FFT_Signal)
plt.plot(IFFT_Signal)
```

```python
y, sr = librosa.load("AI in Speech Processing.mp3", sr=22500)
librosa.display.waveshow(y)
```

```python
sections = librosa.effects.split(y, top_db=10)
```

```python
import soundfile as sf
for i, section in enumerate(sections):
    split_segments = y[section[0]:section[1]]
    sf.write(f'split_segments_{i+1}.mp3',split_segments, 22050)
```

```python
#Chosen section 1 which has the word "AI"
y, sr = librosa.load("split_segments_1.mp3", sr=22500)
split_signal_1 = np.fft.fft(y)
y, sr = librosa.load("split_segments_2.mp3", sr=22500)
split_signal_2 = np.fft.fft(y)
y, sr = librosa.load("split_segments_3.mp3", sr=22500)
split_signal_3 = np.fft.fft(y)
y, sr = librosa.load("split_segments_4.mp3", sr=22500)
split_signal_4 = np.fft.fft(y)
y, sr = librosa.load("AI in Speech Processing.mp3", sr=22500)
signal = np.fft.fft(y)
fig,ax = plt.subplots(5,figsize=(15,25))
librosa.display.waveshow(split_signal_1.real, ax=ax[0])
librosa.display.waveshow(split_signal_2.real, ax=ax[1])
librosa.display.waveshow(split_signal_3.real, ax=ax[2])
librosa.display.waveshow(split_signal_4.real, ax=ax[3])
librosa.display.waveshow(signal.real, ax=ax[4])
ax[0].title.set_text('Split Signal 1')
ax[1].title.set_text('Split Signal 2')
ax[2].title.set_text('Split Signal 3')
ax[3].title.set_text('Split Signal 4')
ax[4].title.set_text('Original Signal')
```

```python
y, sr = librosa.load("AI in Speech Processing.mp3",sr=22500)
librosa.display.waveshow(y)
```

```python
samples_per_section = int(0.02 * sr)
start = int(1.0 * sr)
section = y[start:start+samples_per_section]
librosa.display.waveshow(section)
```

```python
FFT_Segment = np.fft.fft(section)
plt.plot(FFT_Segment.real)
```
